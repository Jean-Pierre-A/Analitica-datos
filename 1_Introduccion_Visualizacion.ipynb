## Manipulación de Datos en Python

```
# Librerías a importar
import pandas as pd # manipular conjuntos de datos (dataframe)
import numpy as np # operaciones matrices
import matplotlib.pyplot as plt # gráficas
```

```
#Lectura de datos
data = pd.read_excel("calidad del aire.xlsx",sheet_name=0)
data.head() #Lista los 5 primeros registros de los datos
```

```
data.info() #Conocer tipos de datos
```

```
#Corrección de tipo de datos
data['alerta']=data['alerta'].astype('category')
data.info()
```

```
data.describe() #Descripción esytadística de las variables numéricas
```

```
#Importamos la librería
import matplotlib.pyplot as plt
%matplotlib inline
```

```
# Diagrama de líneas
x=data['temperatura']
y=data['humedad']
plt.plot(x, y)
plt.xlabel('Nombre del eje X') # definir el nombre del eje X
plt.ylabel('Nombre del eje Y') # definir el nombre del eje Y
plt.title('Titulo de la grafica') # definir el titulo de la grafica
plt.show() # Mostrar la grafica luego de que ya se definio todos los elementos
```

```
# Dos gráficas
plt.subplot(1,2,1) # subplot fila=1 Col=2, grafica=1
plt.plot(x, y, 'r--') # r-- color rojo y linea discontinua
plt.subplot(1,2,2) # subplot fila=1 Col=2, grafica=2
plt.plot(y, x, 'g*-'); # para no mostrar info de la funcion
plt.show() 
```

```
#Dos filas, dos columnas, cambiando los colores
plt.subplot(2,2,1) # subplot fila=1 Col=2, grafica=1
plt.plot(x, y, 'r--') # r-- color rojo y linea discontinua
plt.subplot(2,2,2) # subplot fila=1 Col=2, grafica=2
plt.plot(y, x, 'g*-'); # para no mostrar info de la funcion
plt.subplot(2,2,3) # subplot fila=1 Col=2, grafica=2
plt.plot(y, x, 'y*-'); # para no mostrar info de la funcion
plt.subplot(2,2,4) # subplot fila=1 Col=2, grafica=2
plt.plot(y, x, 'b*-'); # para no mostrar info de la funcion
plt.show() 
```

```
#Dos filas, dos columnas, cambiando los colores
plt.subplot(1,4,1) # subplot fila=1 Col=2, grafica=1
plt.plot(x, y, 'r--') # r-- color rojo y linea discontinua
plt.subplot(1,4,2) # subplot fila=1 Col=2, grafica=2
plt.plot(y, x, 'g*-'); # para no mostrar info de la funcion
plt.subplot(1,4,3) # subplot fila=1 Col=2, grafica=2
plt.plot(y, x, 'y*-'); # para no mostrar info de la funcion
plt.subplot(1,4,4) # subplot fila=1 Col=2, grafica=2
plt.plot(y, x, 'b*-'); # para no mostrar info de la funcion
plt.show() 
```

```
plt.scatter(x, y, s=data['PM25'], c=data['PM25'])
plt.title("Scatter plot");
plt.xlabel('Nombre del eje X') # definir el nombre del eje X
plt.ylabel('Nombre del eje Y') # definir el nombre del eje Y
plt.show()
```

```
plt.hist(data['PM25'])
plt.title("Histograma PM 2.5");
plt.show()
```

```
plt.boxplot(data['PM25']);
plt.title("Boxplot");
plt.show()
```

```
import seaborn as sns
%matplotlib inline 
```

```
#Distribución
sns.distplot(data['PM25']);
```

```
#Gráfica para variable categórica
sns.barplot(x='alerta',y='PM25',data=data);
```

```
#Boxplot para variable categórica
sns.boxplot(x="alerta", y="PM25", data=data,palette='rainbow');
```

```
#Dispersión para variables categóricas
sns.stripplot(x="alerta", y="PM25", data=data);
```

```
#regresión lineal
sns.lmplot(x='temperatura',y='PM25',data=data);
```

```
#Hace una regresión para cada tipo de alerta
sns.lmplot(x='temperatura',y='PM25',data=data,col='alerta') 
```

```
#diagonal histogramas los demas son scatter plots
sns.pairplot(data); 
```

```
#Heatmap de la matriz de correlaciones data.corr()
sns.heatmap(data.corr(),cmap='coolwarm',annot=True);
```

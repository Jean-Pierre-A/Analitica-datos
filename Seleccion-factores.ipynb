Se puede realizar con varios métodos:
 
* Correlaciones
* SelectkBest

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('titanic.csv')
data.info()
```

```python
#Convertimos a número todas las variables (incluyendo la objetivo)
 
#Dummies para variable con más de 2 categorías
data = pd.get_dummies(data, columns=['Clase'], drop_first=False)
 
#Dummies para variable con 2 categorías
data = pd.get_dummies(data, columns=['Edad', 'Sexo','Sobrevivio'],  drop_first=True)
 
data.head()
```

```python
# Correlaciones: variables numéricas 
 
correlaciones=data.corr()
correlaciones
```

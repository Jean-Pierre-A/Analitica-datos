```python
#Importamos librerías básicas

import pandas as pd 

import numpy as np

import matplotlib.pyplot as plt
```

```python
data = pd.read_excel("Datos_Aprendizaje.xlsx")
data.head()
```

```python
data.describe()
```

```python
data.info()
```

```python
data['Figura']=data['Figura'].astype('category')
data.info()
```

```python
from sklearn.preprocessing import LabelEncoder

labelencoder = LabelEncoder()
data['Figura'] = labelencoder.fit_transform(data['Figura'])
data['Figura']
data.head()
```

```python
#División 70-30

from sklearn.model_selection import train_test_split

X = data.drop("Figura", axis = 1) # Variables predictoras

Y = data["Figura"] #Variable objetivo

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, stratify=Y)

Y_train.value_counts().plot(kind="bar")
```

```python
from tensorflow.keras import Sequential

from tensorflow.keras.layers import Dense



#Red neuronal profunda

model_deep = Sequential()

model_deep.add(Dense(50, input_dim=35, activation='sigmoid')) #35 entradas y capa oculta 30 neuronas

model_deep.add(Dense(35, activation='relu')) #capa oculta de 35 neuronas

model_deep.add(Dense(20, activation='relu'))

model_deep.add(Dense(26, activation='softmax')) #capa de salida (valores de la variable objetivo)

model_deep.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy']) 

model_deep.fit(X_train, Y_train, epochs=200)
```

# 4.Evaluacion Sobre el Conjunto

```python
#Evaluación sobre el conjunto de prueba

loss, acc = model_deep.evaluate(X_test, Y_test, verbose = 0)
print(f'Exactitud:{acc}')
```

```python
#Cargamos los datos Futuros
data_fut = pd.read_excel("Datos_Aprendizaje.xlsx", sheet_name="datos futuros")
data_fut.head()
```

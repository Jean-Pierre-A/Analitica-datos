* Preparación: variables categóricas
* Aprendizaje: apriori
* Evaluación: soporte y confianza
* Interpretación: lectura de las reglas}

```python
# Librerías básicas
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_excel('market.xlsx',sheet_name=0)
data.head()
```

```python
#Corrección del tipo de datos
data=data.astype('category')
data.info()
```

```python
#Convertir a datos binarios
data=data.replace({'No': 0, 'Si': 1})
data.head()
```

```python
from mlxtend.frequent_patterns import association_rules, apriori
 
#Productos mas frecuentes
frequent_itemsets = apriori(data, min_support = 0.05, max_len = 2, use_colnames=True)
 
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.01)
rules
```

```python
rules.sort_values("confidence",ascending=False)
rules
```
